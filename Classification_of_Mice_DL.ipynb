## MULTILAYER PERCEPTRONS MODEL (MLP)

```python
# Import Required Libraries
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
from sklearn.model_selection import StratifiedKFold
from sklearn.preprocessing import StandardScaler, LabelEncoder
from sklearn.metrics import accuracy_score, confusion_matrix, ConfusionMatrixDisplay
from tensorflow import keras
from tensorflow.keras import layers, regularizers
from tensorflow.keras.callbacks import EarlyStopping
from tensorflow.keras.utils import to_categorical
import tensorflow as tf

# Reproducibility
np.random.seed(42)
tf.random.set_seed(42)
```

```python
# Load dataset
data = pd.read_csv('Data_Cortex_Nuclear.csv')
data = data.dropna()

# Drop non-numeric/categorical columns from features
non_feature_cols = ['MouseID', 'Genotype', 'Treatment', 'Behavior', 'class']
X = data.drop(columns=non_feature_cols)
y = data['class'].values  # Keep this for classification

# Encode labels
le = LabelEncoder()
y_encoded = le.fit_transform(y)

print("Feature shape:", X.shape)
print("Number of classes:", len(np.unique(y_encoded)))
```

```python
# Define the MLP Model with Regularization
def build_mlp(input_dim, output_dim):
    model = keras.Sequential([
        layers.Input(shape=(input_dim,)),
        
        layers.GaussianNoise(0.1),  # Add stronger noise
        
        layers.Dense(64, activation='relu', kernel_regularizer=regularizers.l2(0.05)),
        layers.BatchNormalization(),
        layers.Dropout(0.5),

        layers.Dense(32, activation='relu', kernel_regularizer=regularizers.l2(0.05)),
        layers.BatchNormalization(),
        layers.Dropout(0.5),

        layers.Dense(output_dim, activation='softmax')
    ])

    lr_schedule = keras.optimizers.schedules.ExponentialDecay(
        initial_learning_rate=0.005,
        decay_steps=100,
        decay_rate=0.9,
        staircase=True
    )

    model.compile(
        optimizer=keras.optimizers.Adam(learning_rate=lr_schedule),
        loss='categorical_crossentropy',
        metrics=['accuracy']
    )
    return model
```

```python
# Stratified K-Fold Cross-Validation with Preprocessing Inside Loop
kfold = StratifiedKFold(n_splits=5, shuffle=True, random_state=42)
val_scores = []

for fold, (train_idx, val_idx) in enumerate(kfold.split(X, y_encoded)):
    # Scale only within the fold
    scaler = StandardScaler()
    X_train = scaler.fit_transform(X.iloc[train_idx])
    X_val = scaler.transform(X.iloc[val_idx])

    # One-hot encode labels
    y_train = to_categorical(y_encoded[train_idx])
    y_val = to_categorical(y_encoded[val_idx])

    # Build and train model
    model = build_mlp(X_train.shape[1], y_train.shape[1])
    early_stop = EarlyStopping(monitor='val_loss', patience=5, restore_best_weights=True)

    history = model.fit(
        X_train, y_train,
        validation_data=(X_val, y_val),
        epochs=50,
        batch_size=32,
        verbose=1,
        callbacks=[early_stop]
    )

    # Evaluate
    val_loss, val_acc = model.evaluate(X_val, y_val, verbose=0)
    val_scores.append(val_acc)
    print(f"Fold {fold+1} accuracy: {val_acc:.4f}")

print("\nCross-validation accuracy scores:", val_scores)
print(f"Mean accuracy: {np.mean(val_scores):.4f}")
```

```python
# importing libraries
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
from sklearn.model_selection import train_test_split
from sklearn.preprocessing import LabelEncoder, StandardScaler
from sklearn.impute import SimpleImputer
from sklearn.metrics import confusion_matrix, ConfusionMatrixDisplay
from tensorflow.keras.models import Sequential
from tensorflow.keras.layers import Conv2D, MaxPooling2D, Flatten, Dense, Dropout, BatchNormalization, LeakyReLU
from tensorflow.keras.utils import to_categorical
from tensorflow.keras.callbacks import EarlyStopping
import tensorflow as tf
```

```python
# Set random seeds for reproducibility
np.random.seed(42)
tf.random.set_seed(42)

# Load dataset
df = pd.read_csv("Data_Cortex_Nuclear.csv")
df.drop(['Genotype', 'Treatment', 'Behavior', 'MouseID'], axis=1, inplace=True)

# Encode target labels
label_encoder = LabelEncoder()
df['class'] = label_encoder.fit_transform(df['class'])

# Split features and target
X = df.drop('class', axis=1).values.astype(np.float32)
y = df['class'].values

```

```python
# Check for NaNs/Infs in raw data
print("NaNs in X before cleaning:", np.isnan(X).any())
print("Infs in X before cleaning:", np.isinf(X).any())

# Clean NaNs/Infs using imputation
imputer = SimpleImputer(strategy="mean")
X = imputer.fit_transform(X)

# Double-check after cleaning
assert not np.isnan(X).any(), "NaNs still present in features!"
assert not np.isinf(X).any(), "Infs still present in features!"
```

```python
# Scale features
scaler = StandardScaler()
X_scaled = scaler.fit_transform(X)

# Pad/truncate to 81 features
if X_scaled.shape[1] < 81:
    X_padded = np.pad(X_scaled, ((0, 0), (0, 81 - X_scaled.shape[1])), mode='constant')
else:
    X_padded = X_scaled[:, :81]

# Reshape into 9x9x1 "image"
X_reshaped = X_padded.reshape(-1, 9, 9, 1)
```

```python
# preprocess and reshaping

# One-hot encode labels
y_cat = to_categorical(y)

# Final check
assert not np.isnan(X_reshaped).any(), "NaNs found in reshaped X"
assert not np.isinf(X_reshaped).any(), "Infs found in reshaped X"
assert not np.isnan(y_cat).any(), "NaNs found in y_cat"

# Train-test split
X_train, X_test, y_train, y_test = train_test_split(
    X_reshaped, y_cat, test_size=0.3, random_state=42, stratify=y_cat
)
```

```python
# Add noise to training data
noise_factor = 0.1
X_train_noisy = X_train + noise_factor * np.random.normal(loc=0.0, scale=1.0, size=X_train.shape)
X_train_noisy = np.clip(X_train_noisy, -1., 1.)

# Updated CNN Model with regularization and increased dropout
model = Sequential([
    Conv2D(16, (3, 3), padding='same', input_shape=(9, 9, 1), use_bias=False,
           kernel_initializer='he_normal', kernel_regularizer=regularizers.l2(0.01)),
    BatchNormalization(),
    LeakyReLU(),
    MaxPooling2D((2, 2)),
    Dropout(0.3),

    Conv2D(32, (3, 3), padding='same', use_bias=False,
           kernel_initializer='he_normal', kernel_regularizer=regularizers.l2(0.01)),
    BatchNormalization(),
    LeakyReLU(),
    MaxPooling2D((2, 2)),
    Dropout(0.4),

    Flatten(),

    Dense(64, use_bias=False, kernel_initializer='he_normal',
          kernel_regularizer=regularizers.l2(0.01)),
    BatchNormalization(),
    LeakyReLU(),
    Dropout(0.4),

    Dense(32, use_bias=False, kernel_initializer='he_normal',
          kernel_regularizer=regularizers.l2(0.01)),
    BatchNormalization(),
    LeakyReLU(),
    Dropout(0.3),

    Dense(y_cat.shape[1], activation='softmax')
])

model.compile(optimizer='adam', loss='categorical_crossentropy', metrics=['accuracy'])

# Summary
model.summary()
```

```python
# Early stopping
early_stop = EarlyStopping(monitor='val_loss', patience=15, restore_best_weights=True)

# Training
history = model.fit(
    X_train_noisy, y_train,
    validation_split=0.2,
    epochs=100,
    batch_size=16,
    callbacks=[early_stop],
    verbose=1
)
```

```python
# Evaluation
loss, accuracy = model.evaluate(X_test, y_test)
print(f"Test Accuracy: {accuracy:.4f}")
```
